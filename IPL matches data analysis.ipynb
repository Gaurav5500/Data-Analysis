## Most wins in IPL

```python
wins = pd.DataFrame({"Winner": data['winner']})
```

```python
count_wins = wins.value_counts()
```

```python
print(count_wins)
```

```python
labels = [X[0]for X in count_wins.keys()]
bar, ax = plt.subplots(figsize=(18,12))
ax = plt.pie(x=count_wins, autopct='%.1f%%',labels = labels)
plt.title("Most wins in IPL",fontsize=18)
plt.show()
```

## Most Wins In Eliminator

```python
sns.countplot(data['winner'][data['eliminator']=='Y'],data=data)
plt.title("Most wins in the Eliminator matches",fontsize=18)
plt.xticks(rotation=35)
plt.show()
```

## Toss Decision

```python
sns.countplot(y=data['toss_winner'])
plt.xticks(rotation=90)
plt.plot()
```

## Famous Venue

```python
sns.barplot(x=data['venue'].value_counts().head(8).values,
            y=data['venue'].value_counts().head(8).index, data=data)
plt.title("Famous Venues")
plt.xlabel("Venue Count")
plt.ylabel("Venue")
```

## Top 5 Umpires

```python
sns.barplot(x=data['umpire1'].value_counts().head().index,
            y=data['umpire1'].value_counts().head().values, data=data)
plt.xlabel("Umpire1")
plt.ylabel("MartchCount")
plt.title('Top Umpires')
plt.xticks(rotation=35)
plt.show()
```

```python
sns.barplot(x=data['umpire2'].value_counts().head().values,
            y=data['umpire2'].value_counts().head().index, data=data)
plt.ylabel("Umpire2")
plt.xlabel("MartchCount")
plt.title('Top Umpires')
plt.xticks(rotation=35)
plt.show()
```
